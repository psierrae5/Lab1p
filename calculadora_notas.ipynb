```python
import random
puntuaciones = [random.randint(1, 100) for i in range(0,5)]
```

```python
puntuaciones
```

```python
def calcular_promedio_puntuaciones (puntuaciones:list):
    suma=0
    for i in puntuaciones:
        suma+=i
    return suma/len(puntuaciones)
```

```python
def calcular_puntuacion_maxima (puntuaciones:list):
    maximo = 0
    for i in puntuaciones:
        if i > maximo:
            maximo = i
    return maximo
```

```python
def calcular_total_puntuaciones (puntuaciones:list):
    suma=0
    for i in puntuaciones:
        suma+=i
    return suma
```

```python
print(calcular_promedio_puntuaciones(puntuaciones))
```

```python
print(calcular_puntuacion_maxima(puntuaciones))
```

```python
print(calcular_total_puntuaciones(puntuaciones))
```
